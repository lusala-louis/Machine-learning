```python
import requests
from bs4 import BeautifulSoup

def scrape_book_titles(url):
    """Scrape book titles from the given URL."""
    titles = []
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        book_elements = soup.select('.entry-content li')  # CSS selector for book titles
        for element in book_elements:
            title = element.text.strip()
            titles.append(title)
    return titles

url = "https://libraryof1000books.wordpress.com/the-list-of-1000-books/"
book_titles = scrape_book_titles(url)

# Print the list of book titles
for title in book_titles:
    print(title)
```
