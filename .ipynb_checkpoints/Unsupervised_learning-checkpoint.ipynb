# Iris Dimensionality using Principal Component Analysis
Reducing the dimensionality of the iris dta so that we can easily visualize it.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import datasets
```

```python
#Loading the data
iris = datasets.load_iris()

#Conveting the dataset into a dataframe
iris = pd.DataFrame(data = np.c_[iris['data'], iris['target']], columns = iris['feature_names'] + ['species'])

#Replacing the values in the species column with species names
iris.species = np.where(iris.species == 0.0, 'setosa', np.where(iris.species == 1.0, 'versicolor', 'virginica'))

iris.head()
```

```python
#Creating the feature matrix and the target array
X_iris = iris.drop('species', axis = 1)        #Feature matrix

y_iris = iris['species']                       #Target array
```

```python
from sklearn.decomposition import PCA #1. Choose model class
model = PCA(n_components = 2)         #2. Instantiate the model with hyperparameters
model.fit(X_iris)                     #3. Fit to data and in this case y is not specified
X_2D = model.transform(X_iris)        #4. Transform the data to two dimensions
```

```python
#Plotting the results
import seaborn as sns
iris['PCA1'] = X_2D[:, 0]
iris['PCA2'] = X_2D[:, 1]

sns.lmplot("PCA1", "PCA2", hue = 'species', data = iris, fit_reg = False);
```

## Iris Clustering using Gaussian Mixture Models

```python

from sklearn import mixture   #1. Choose model class
from sklearn.mixture import GaussianMixture

model = GaussianMixture(n_components = 3, covariance_type = 'full')#Instantiate the model with hyperparameters
model.fit(X_iris)                             #3. Fit model to the data 
y_gmm = model.predict(X_iris)                 #4. Determine cluster labels
```

```python
#Add the cluster label to the iris data and use seaborn to plot the data
iris['cluster'] = y_gmm
sns.lmplot("PCA1", "PCA2", data = iris, hue = 'species', col = 'cluster', fit_reg =False);
```
