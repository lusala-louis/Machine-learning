```python
import pandas as pd
counts = pd.read_csv('Fremont_Bridge_Bicycle_Counter.csv', index_col = 'Date', parse_dates=True)
weather = pd.read_csv('2876205.csv', index_col='DATE', parse_dates=True)
```

```python
#Compute the total daily traffic 
daily = counts.resample('d').sum()
daily['Total'] = daily.sum(axis=1)
daily = daily[['Total']]                #Removes other columns
daily.head()
```

```python
#Adding binary columns that add the days of the week
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
for i in range(7):
    daily[days[i]] = (daily.index.dayofweek == i).astype(float)

daily.head()
```

```python
#Adding an indicator for holidays
from pandas.tseries.holiday import USFederalHolidayCalendar
cal = USFederalHolidayCalendar()
holidays = cal.holidays('2012', '2016')
daily = daily.join(pd.Series(1, index = holidays, name = 'holiday'))
daily['holiday'].fillna(0, inplace =True)
daily.head()
```

```python
#Adding a condition for hours of daylight
import numpy as np
def hours_of_daylight(date, axis=23.44, latitude=47.61):
    """Compute the hours of daylight for the given data"""
    days = (date - pd.datetime(2000, 12, 21)).days
    m = (1 - np.tan(np.radians(latitude)) 
         * np.tan(np.radians(axis) * np.cos(days *2 * np.pi / 365.25)))
    return 24. * np.degrees(np.arccos(1 - np.clip(m, 0, 2)))/ 100.

daily['dailylight_hrs'] = list(map(hours_of_daylight, daily.index))
daily[['dailylight_hrs']].plot();
```
