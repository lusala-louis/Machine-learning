# Univariate Data Analysis
Individual variables are analyzed in isolation to have a better understanding about them

```python
from sklearn import datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()

#Converting the data into a dataframe
iris = pd.DataFrame(data = np.c_[iris['data'], iris['target']], columns = iris['feature_names'] + ['species'])
iris.head(10)
```

```python
#Replace the values with class names
iris.species = np.where(iris.species == 0.0, 'setosa', np.where(iris.species == 1.0, 'versicolor','virginica'))
iris
```

```python
#Remove spaces from column name
iris.columns = iris.columns.str.replace(' ', '')
iris
```

```python
#Getting summary statistics for the variables
iris.describe()
```

```python
#Frequency distribution for each category in species column
iris['species'].value_counts()
```

## Visualizing the data using matplotlib

```python
#Setting the size of the plot
plt.figure(figsize=(15,10), dpi = 300)

iris.hist()                              #Plots a histogram
plt.suptitle("Histogram", fontsize = 16) #Adds one subtitle to all
plt.show()

iris.boxplot()                           #Plots a boxplot
plt.title("Bar Plot", fontsize=16)
plt.show()
```
